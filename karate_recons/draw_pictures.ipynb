```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from torch_geometric.datasets import KarateClub
```

```python
nodes_number = 34
dataset = KarateClub()
data = dataset[0]
edges = data.edge_index
start_nodes = np.array(edges[0])
end_nodes = np.array(edges[1])
graph_np = np.zeros((nodes_number, nodes_number))
for i in range(edges.shape[1]):
    graph_np[start_nodes[i], end_nodes[i]]=1
G_true=nx.from_numpy_array(graph_np)
```

```python
ClusterLP_graph = pd.read_csv("ClusterLP_recon_karate.txt", header=None, sep=' ')
ClusterLP_graph = np.array(ClusterLP_graph)
ClusterLP_predict = nx.from_numpy_array(ClusterLP_graph)
```

```python
LINE_graph = pd.read_csv("LINE_recon_karate.txt", header=None, sep=' ')
LINE_graph = np.array(LINE_graph)
LINE_predict = nx.from_numpy_array(LINE_graph)

NODE2VEC_graph = pd.read_csv("NODE2VEC_recon_karate.txt", header=None, sep=' ')
NODE2VEC_graph = np.array(NODE2VEC_graph)
NODE2VEC_predict = nx.from_numpy_array(NODE2VEC_graph)

VGAE_graph = pd.read_csv("VGAE_recon_karate.txt", header=None).iloc[:,1:]
VGAE_graph = np.array(VGAE_graph)
VGAE_predict = nx.from_numpy_array(VGAE_graph)

AGE_graph = pd.read_csv("AGE_recon_karate.txt", header=None, sep=' ')
AGE_graph = np.array(AGE_graph)
AGE_predict = nx.from_numpy_array(AGE_graph)

ARVGA_graph = pd.read_csv("ARVGA_recon_karate.txt", header=None, sep=' ')
ARVGA_graph = np.array(ARVGA_graph)
ARVGA_predict = nx.from_numpy_array(ARVGA_graph)

NECS_graph = pd.read_csv("NECS_recon_karate.txt", header=None, sep=' ')
NECS_graph = np.array(NECS_graph)
NECS_predict = nx.from_numpy_array(NECS_graph)

GIC_graph = pd.read_csv("GIC_recon_karate.txt", header=None, sep=' ')
GIC_graph = np.array(GIC_graph)
GIC_predict = nx.from_numpy_array(GIC_graph)

LMA_graph = pd.read_csv("LMA_recon_karate.txt", header=None, sep=' ')
LMA_graph = np.array(LMA_graph)
LMA_predict = nx.from_numpy_array(LMA_graph)

NAFS_graph = pd.read_csv("NAFS_recon_karate.txt", header=None, sep=' ')
NAFS_graph = np.array(NAFS_graph)
NAFS_predict = nx.from_numpy_array(NAFS_graph)
```

```python
plt.figure(figsize=(27, 40))
plt.subplots_adjust(wspace=0.05, hspace=0.05)

grid=plt.GridSpec(4, 6)

AX_TRUE = plt.subplot(grid[0, 1:3])
true_pos = nx.kamada_kawai_layout(G_true)
nx.draw(G_true,true_pos,with_labels = True,node_color = 'pink',node_size = 400, width=1, ax=AX_TRUE)
plt.title(r"$Karate\ Network$", fontsize=30, y=-0.03)

AX_CLUSTERLP = plt.subplot(grid[0, 3:5])
ClusterLP_pos=nx.kamada_kawai_layout(ClusterLP_predict)
nx.draw(ClusterLP_predict, ClusterLP_pos, with_labels=True, node_color='peachpuff', node_size = 400, width=1, ax=AX_CLUSTERLP)
plt.title(r"$ClusterLP$", fontsize=30, y=-0.03)

AX0 = plt.subplot(grid[1, 0:2])
LINE_pos = nx.kamada_kawai_layout(LINE_predict)
nx.draw(LINE_predict, LINE_pos, with_labels = True, node_color = 'c', node_size = 300, width=1, ax=AX0)
plt.title(r"$LINE$", fontsize=30, y=-0.03)

AX1 = plt.subplot(grid[1, 2:4])
NODE2VEC_pos = nx.kamada_kawai_layout(NODE2VEC_predict)
nx.draw(NODE2VEC_predict, NODE2VEC_pos, with_labels = True, node_color = 'coral', node_size = 300, width=1, ax=AX1)
plt.title(r"$NODE2VEC$", fontsize=30, y=-0.03)

AX2 = plt.subplot(grid[1, 4:])
NECS_pos = nx.kamada_kawai_layout(NECS_predict)
nx.draw(NECS_predict, NECS_pos, with_labels = True, node_color = 'orange', node_size = 300, width=1, ax=AX2)
plt.title(r"$NECS$", fontsize=30, y=-0.03)

AX3 = plt.subplot(grid[2, 0:2])
VGAE_pos = nx.kamada_kawai_layout(VGAE_predict)
nx.draw(VGAE_predict, VGAE_pos, with_labels = True, node_color = 'pink', node_size = 300, width=1, ax=AX3)
plt.title(r"$VGAE$", fontsize=30, y=-0.03)

AX4 = plt.subplot(grid[2, 2:4])
ARVGA_pos = nx.kamada_kawai_layout(ARVGA_predict)
nx.draw(ARVGA_predict, ARVGA_pos, with_labels = True, node_color = 'limegreen', node_size = 300, width=1, ax=AX4)
plt.title(r"$ARVGA$", fontsize=30, y=-0.03)

AX5 = plt.subplot(grid[2, 4:])
AGE_pos = nx.kamada_kawai_layout(AGE_predict)
nx.draw(AGE_predict, AGE_pos, with_labels = True, node_color = 'gold', node_size = 300, width=1, ax=AX5)
plt.title(r"$AGE$", fontsize=30, y=-0.03)

AX6 = plt.subplot(grid[3, 0:2])
GIC_pos = nx.kamada_kawai_layout(GIC_predict)
nx.draw(GIC_predict, GIC_pos, with_labels = True, node_color = 'plum', node_size = 300, width=1, ax=AX6)
plt.title(r"$GIC$", fontsize=30, y=-0.03)

AX7 = plt.subplot(grid[3, 2:4])
LMA_pos = nx.kamada_kawai_layout(LMA_predict)
nx.draw(LMA_predict, LMA_pos, with_labels = True, node_color = 'tomato', node_size = 300, width=1, ax=AX7)
plt.title(r"$LMA$", fontsize=30, y=-0.03)

AX8 = plt.subplot(grid[3, 4:6])
NAFS_pos = nx.kamada_kawai_layout(NAFS_predict)
nx.draw(NAFS_predict, NAFS_pos, with_labels = True, node_color = 'peachpuff', node_size = 300, width=1, ax=AX8)
plt.title(r"$NAFS$", fontsize=30, y=-0.03)

plt.savefig("baselines_reconed_karate.pdf",bbox_inches='tight')
plt.show()
```
